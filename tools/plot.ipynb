```python
import matplotlib.pyplot as plt

# 数据点
x = [10,    20,     30,     40,     100]
y = [8.5,   13.2,   18.1,   22.8,   51.2]

# 绘制折线图
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Example Line')

# 添加标签和标题
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.legend()

# 显示图形
plt.grid(True)
plt.show()
```

```python
import numpy as np
import time


start=time.time()

tmp=np.ones(8*(1024**3))
res=np.empty_like(tmp)

mid=time.time()

np.multiply(tmp, 2, out=res)

end=time.time()
print(mid-start)
print(end-mid)
```

```python
import multiprocessing as mp
import time

chunk_len = 2**31  # 每个 float64 = 8 bytes => 16 GB

start = time.time()
shared_arr = mp.RawArray('d', chunk_len, lock=False)
end = time.time()

print(f"Allocated 16 GB shared memory in {end - start:.2f} seconds.")
```
